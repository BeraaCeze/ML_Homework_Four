```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("polynomial-regression.csv",sep = ";")

# ( car_max_speed ) features
# ( car_price ) output

x = df.car_max_speed.values.reshape(-1,1)  # get featur (car_max_speed)
y = df.car_price.values.reshape(-1,1)      # get output (car_price)

plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")
plt.show()
```

```python
# linear regression =  y = b0 + b1*x
# multiple linear regression   y = b0 + b1*x1 + b2*x2

# %% linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()  # create object from LinearRegression model

lr.fit(x,y)              # make fit

#%% predict x
y_head = lr.predict(x)   
  
plt.scatter(x,y , color="green")
plt.xlabel("car_max_speed")
plt.ylabel("car_price")

plt.plot(x,y_head,color="red",label ="linear")
plt.show()

print("10 million TL car speed estimation: ",lr.predict([[1000]]))
```

```python
# %%
# polynomial regression =  y = b0 + b1*x +b2*x^2 + b3*x^3 + ... + bn*x^n

print (x.shape)  

from sklearn.preprocessing import PolynomialFeatures
polynomial_regression = PolynomialFeatures(degree = 2)  # increase 2 feature  


x_polynomial = polynomial_regression.fit_transform(x)

print (x_polynomial.shape)


# %% fit
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)     #  train with new features

# %%

y_head2 = linear_regression2.predict(x_polynomial)

plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")
plt.plot(x,y_head,color="red",label ="linear")     # old regression

plt.plot(x,y_head2,color= "green",label = "poly")  # new regression is more accuracy
plt.legend()
plt.show()

```

```python
#%%
from sklearn.metrics import r2_score  

print("r_square score for linear regression: ", r2_score(y,y_head))

print("r_square score for polynomial regression: ", r2_score(y,y_head2))
```
